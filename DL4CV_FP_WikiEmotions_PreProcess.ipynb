```python
import pandas as pd
import os

from PIL import Image
import requests
from io import BytesIO
import shutil
```

# To parse and download all the WikiArt-Emotions images

```python
info = pd.read_csv('/content/drive/MyDrive/Data Science - Master Notes/DL4CV Final Project/WikiArt-info.tsv', sep='\t')
```

```python
total = len(info)
def update(ind):
  if ind % 250 == 0:
    print(ind, '/', total)
```

```python
newpath = '/content/WikiArt-imgs' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
```

```python
def img_export(x):
  id = x['ID']
  url = x['Image URL']
  update(x.name)
  try:
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))
    img.save("/content/WikiArt-imgs/" + id + ".png")
  except:
    print("Error on,")
    print("   id: ", id)
    print("   url: ", url)
    print()
```

```python
info.apply(img_export, axis=1)
```

# Divide into train, val, test by class

```python
wiki_em_labels = pd.read_csv('/content/wiki_em_labels.csv')
```

```python
wiki_em_labels = wiki_em_labels[wiki_em_labels['ID'] != '59839e9cedc2c9658c2f54cd']
```

```python
pos = wiki_em_labels[wiki_em_labels['label'] == 'positive']
neg = wiki_em_labels[wiki_em_labels['label'] == 'negative']
other = wiki_em_labels[wiki_em_labels['label'] == 'other']

pos_train = pos.sample(frac = 0.8)
pos_test = pos.drop(pos_train.index)
pos_val = pos_train.sample(frac = 0.1)
pos_train = pos_train.drop(pos_val.index)

neg_train = neg.sample(frac = 0.8)
neg_test = neg.drop(neg_train.index)
neg_val = neg_train.sample(frac = 0.1)
neg_train = neg_train.drop(neg_val.index)

other_train = other.sample(frac = 0.8)
other_test = other.drop(other_train.index)
other_val = other_train.sample(frac = 0.1)
other_train = other_train.drop(other_val.index)

pos_train = pos_train.to_dict('records')
pos_val = pos_val.to_dict('records')
pos_test = pos_test.to_dict('records')

neg_train = neg_train.to_dict('records')
neg_val = neg_val.to_dict('records')
neg_test = neg_test.to_dict('records')

other_train = other_train.to_dict('records')
other_val = other_val.to_dict('records')
other_test = other_test.to_dict('records')
```

```python
newpath = '/content/data' 
if not os.path.exists(newpath):
    os.makedirs(newpath)

newpath = '/content/data/train' 
if not os.path.exists(newpath):
    os.makedirs(newpath)

newpath = '/content/data/train/positive' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
newpath = '/content/data/train/negative' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
newpath = '/content/data/train/other' 
if not os.path.exists(newpath):
    os.makedirs(newpath)

newpath = '/content/data/validation' 
if not os.path.exists(newpath):
    os.makedirs(newpath)

newpath = '/content/data/validation/positive' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
newpath = '/content/data/validation/negative' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
newpath = '/content/data/validation/other' 
if not os.path.exists(newpath):
    os.makedirs(newpath)

newpath = '/content/data/test' 
if not os.path.exists(newpath):
    os.makedirs(newpath)

newpath = '/content/data/test/positive' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
newpath = '/content/data/test/negative' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
newpath = '/content/data/test/other' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
```

```python
def copy_file(current_path, new_path):
  if os.path.isfile(current_path):
    shutil.copy(current_path, new_path)
```

```python
for item in pos_train:
  id = item['ID']
  current_path = '/content/WikiArt-imgs/' + id + '.png'
  new_path = '/content/data/train/positive/' + id + '.png'
  copy_file(current_path, new_path)
for item in neg_train:
  id = item['ID']
  current_path = '/content/WikiArt-imgs/' + id + '.png'
  new_path = '/content/data/train/negative/' + id + '.png'
  copy_file(current_path, new_path)
for item in other_train:
  id = item['ID']
  current_path = '/content/WikiArt-imgs/' + id + '.png'
  new_path = '/content/data/train/other/' + id + '.png'
  copy_file(current_path, new_path)
```

```python
for item in pos_val:
  id = item['ID']
  current_path = '/content/WikiArt-imgs/' + id + '.png'
  new_path = '/content/data/validation/positive/' + id + '.png'
  copy_file(current_path, new_path)
for item in neg_val:
  id = item['ID']
  current_path = '/content/WikiArt-imgs/' + id + '.png'
  new_path = '/content/data/validation/negative/' + id + '.png'
  copy_file(current_path, new_path)
for item in other_val:
  id = item['ID']
  current_path = '/content/WikiArt-imgs/' + id + '.png'
  new_path = '/content/data/validation/other/' + id + '.png'
  copy_file(current_path, new_path)
```

```python
for item in pos_test:
  id = item['ID']
  current_path = '/content/WikiArt-imgs/' + id + '.png'
  new_path = '/content/data/test/positive/' + id + '.png'
  copy_file(current_path, new_path)
for item in neg_test:
  id = item['ID']
  current_path = '/content/WikiArt-imgs/' + id + '.png'
  new_path = '/content/data/test/negative/' + id + '.png'
  copy_file(current_path, new_path)
for item in other_test:
  id = item['ID']
  current_path = '/content/WikiArt-imgs/' + id + '.png'
  new_path = '/content/data/test/other/' + id + '.png'
  copy_file(current_path, new_path)
```
