```python
import pandas as pd
import shutil
import os
from sklearn.model_selection import train_test_split
```

```python
shutil.unpack_archive("/content/drive/MyDrive/Data Science - Master Notes/DL4CV Final Project/emotion_dataset.tar", "emotion_dataset")
```

```python
amusement_files = os.listdir('/content/emotion_dataset/emotion_dataset/amusement')
anger_files = os.listdir('/content/emotion_dataset/emotion_dataset/anger')
awe_files = os.listdir('/content/emotion_dataset/emotion_dataset/awe')
contentment_files = os.listdir('/content/emotion_dataset/emotion_dataset/contentment')
disgust_files = os.listdir('/content/emotion_dataset/emotion_dataset/disgust')
excitement_files = os.listdir('/content/emotion_dataset/emotion_dataset/excitement')
fear_files = os.listdir('/content/emotion_dataset/emotion_dataset/fear')
sadness_files = os.listdir('/content/emotion_dataset/emotion_dataset/sadness')
```

```python
amusement_files = ['/content/emotion_dataset/emotion_dataset/amusement/' + x for x in amusement_files]
anger_files = ['/content/emotion_dataset/emotion_dataset/anger/' + x for x in anger_files]
awe_files = ['/content/emotion_dataset/emotion_dataset/awe/' + x for x in awe_files]
contentment_files = ['/content/emotion_dataset/emotion_dataset/contentment/' + x for x in contentment_files]
disgust_files = ['/content/emotion_dataset/emotion_dataset/disgust/' + x for x in disgust_files]
excitement_files = ['/content/emotion_dataset/emotion_dataset/excitement' + x for x in excitement_files]
fear_files = ['/content/emotion_dataset/emotion_dataset/fear/' + x for x in fear_files]
sadness_files = ['/content/emotion_dataset/emotion_dataset/sadness/' + x for x in sadness_files]
```

```python
# Accordingn to paper
pos = amusement_files + awe_files + contentment_files + excitement_files
neg = anger_files + disgust_files + fear_files + sadness_files
```

```python
pos_dev, pos_test, trash, trash = train_test_split(
    pos, [i for i in range(0, len(pos))], test_size=0.33, random_state=42)
```

```python
neg_dev, neg_test, trash, trash = train_test_split(
    neg, [i for i in range(0, len(neg))], test_size=0.33, random_state=42)
```

```python
pos_train, pos_val, trash, trash = train_test_split(
    pos_dev, [i for i in range(0, len(pos_dev))], test_size=0.33, random_state=42)
```

```python
neg_train, neg_val, trash, trash = train_test_split(
    neg_dev, [i for i in range(0, len(neg_dev))], test_size=0.33, random_state=42)
```

```python
def copy_file(current_path, new_path):
  if os.path.isfile(current_path):
    shutil.copy(current_path, new_path)
```

```python
for item in pos_train:
  new_path = '/content/data/train/positive/' + item.split("/")[0]
  copy_file(item, new_path)
for item in neg_train:
  new_path = '/content/data/train/negative/' + item.split("/")[0]
  copy_file(item, new_path)
```

```python
for item in pos_val:
  new_path = '/content/data/validation/positive/' + item.split("/")[0]
  copy_file(item, new_path)
for item in neg_val:
  new_path = '/content/data/validation/negative/' + item.split("/")[0]
  copy_file(item, new_path)
```

```python
for item in pos_test:
  new_path = '/content/data/test/positive/' + item.split("/")[0]
  copy_file(item, new_path)
for item in neg_test:
  new_path = '/content/data/test/negative/' + item.split("/")[0]
  copy_file(item, new_path)
```
